```python
stops = stopwords.words('english')
porter = PorterStemmer()
lemma = WordNetLemmatizer()

tbl = dict.fromkeys(i for i in range(sys.maxunicode)
                      if unicodedata.category(chr(i)).startswith('P') or i == 36 or i == ord('`'))


def remove_punctuation(text):
    '''
    From https://stackoverflow.com/questions/11066400/remove-punctuation-from-unicode-formatted-strings

    '''
    return text.translate(tbl)


def remove_stopwords(text, ret_format='str'):
    tokens = filter(lambda x: x not in stops, map(porter.stem, word_tokenize(text)))
    if ret_format == 'list':
        return list(tokens)
    elif ret_format == 'str':
        return ' '.join(tokens)
    else:
        raise Exception('Invalid format')


def restore_arr(a):
    '''
        Converts strings to python list
        
        params:
            a: String -> Input string to be converted to array
        return:
            list
        
        Usage with pandas:
            train_mod = pd.read_csv('modified_train.csv', converters={'description_norm': restore_arr})
    '''
    return [x.replace("'", "") for x in a[:-1][1:].split(', ')]
        
    

def restore_int_arr(a):
    return [int(x.replace("'", "")) for x in a[:-1][1:].split(', ')]
    

def restore_float_arr(a):
    ret = [float(x.replace("'", "")) for x in a[:-1][1:].split(', ')]
    if len(ret) == 1:
        return ret[0]
    else:
        return ret
    
def lemmatize(a):
    return [lemma.lemmatize(x) for x in a.split()]

def remove_numbers(a):
    ans = []
    for s in a.split():
        try:
            g = int(s)
        except ValueError:
            ans.append(s)
            
    return ' '.join(ans)
```

```python
params = {
    'vocab_size': 2000,
    'PAD': 0,
    'UNK': 1,
    'maxlen': df['text_norm'].map(len).max(),
    'kernels': (3, 3, 3,),
    'num_filters': (2, 2, 2),
    'hidden_dims': 64,
    'batch_size': 32,
    'embedding_size': 64,
    'pool_size': 2,
    'threshold': 0.5,
    'LSTM_units': 64
}
```

```python
vocabulary = dict()
counter = Counter()

for sent in train['text_norm']:
    counter.update(sent)
    
display("10 most common words")
display(counter.most_common(10)) 
display("Number of unique tokens")
display(len(counter.keys()))
```

```python
unk_count = 0
pad_count = 0

def convert_to_dict(arr, maxlen, dictionary, pad=True):
    global unk_count, pad_count
    
    if pad:
#         ret = np.full(shape=(maxlen, ), fill_value=params['PAD'])
        ret = [params['PAD'] for x in range(maxlen)]
    else:
        ret = np.full(shape=(len(arr) + 1, ), fill_value=params['PAD'])
        
    for i, word in enumerate(arr):
        if word in dictionary:
            ret[i] = dictionary[word]
        else:
            ret[i] = params['UNK']
            unk_count += 1
            
    pad_count += maxlen - len(arr)
    return ret
    

def create_dataset(cnt, train):
    global pad_count, unk_count
    
    count = [['PAD', -2], ['UNK', -1]]
    count.extend(cnt.most_common(params['vocab_size'] - 2))
    dictionary = dict()
    for word, _ in count:
        dictionary[word] = len(dictionary)
    
    reverse_dictionary = dict(zip(dictionary.values(), dictionary.keys()))
    series = train['text_norm'].apply(convert_to_dict, args=(params['maxlen'], dictionary))
    count[0][-1] = pad_count
    count[1][-1] = unk_count
    return series, dictionary, reverse_dictionary, count

train['description_vectors'], word_dict, word_rev_dict, count = create_dataset(counter, train)
print('Most common words + (PAD and UNK) ', count[:8])
print('Dataset','\n', train['description_vectors'][:5])
```

```python
desc = np.array([np.array(x) for x in train['description_vectors']])

x_train, x_test, y_train, y_test = train_test_split(
                                        desc, 
                                        train['Difference'], 
                                        test_size=0.1, 
                                        random_state=4)

print(x_train.shape, x_test.shape, y_train.shape)
```

```python
convolutions = list()

input_shape = (params['maxlen'], )
conv_input = Input(shape=input_shape)

embed = Embedding(params['vocab_size'],
                  params['embedding_size'],
                  input_length=params['maxlen'],
                  name='Embedding'
                 )(conv_input)


embed = Dropout(0.5, name='Dropout_1')(embed)

for i, size in enumerate(params['kernels']):
    conv = Conv1D(filters=params['num_filters'][i], kernel_size=size, padding='same', activation='relu')(embed)
    conv = MaxPool1D(pool_size=params['pool_size'])(conv)
    conv = Flatten()(conv)
    convolutions.append(conv)
    
out = concatenate(convolutions)
out = Dropout(0.4, name='Dropout_2')(out)
out = Dense(params['hidden_dims'], activation='relu', name='Dense_hidden')(out)
out = Dense(1, activation='softmax', name='Dense_final')(out)

model = Model(inputs=conv_input, outputs=out)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, 
          batch_size=params['batch_size'], 
          validation_data=(x_test, y_test),
          verbose=1, epochs=3, callbacks=[ModelCheckpoint('model_cnn_v1.keras', save_best_only=True)])
```
